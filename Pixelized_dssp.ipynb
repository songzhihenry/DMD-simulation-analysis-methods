```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
```

# Load dssp file

```python
dssp=[line.strip('\n') for line in open('sample.dssp')]
```

# Define a color map

## String to Secondary structures:
- 'H' = alpha-helix
- 'B' = isolated beta bridge
- 'E' = beta-sheet
- 'G' = 3-helix
- 'I' = 5-helix
- 'T' = H-bonded turn
- 'S' = bend
- 'C' = coil

![SS & Colors](ss_pixil_colors.png)

```python
char_to_color = {'C':(0.5, '#ffffff'),
                 'E':(1.5, '#e61f18'),
                 'B':(2.5, '#040000'),
                 'S':(3.5, '#127c3c'),
                 'T':(4.5, '#efe93b'),
                 'H':(5.5, '#18489c'),
                 'I':(6.5, '#761c76'),
                 'G':(7.5, '#7e7e7e')}
```

# Convert dssp to array

```python
# interval is set to be 100 to avoid a large figure
pixel_data = np.array([[char_to_color[char][0] for char in line] for line in dssp[::100]]).T
```

# Make a color map of fixed colors

```python
cmap = colors.ListedColormap([hex[1] for _,hex in char_to_color.items()])
bounds = [hex[0] for _,hex in char_to_color.items()]
norm = colors.BoundaryNorm(bounds, cmap.N)
```

# Plot SS

```python
plt.figure(figsize=(20,5))
img = plt.imshow(pixel_data, origin='lower',cmap=cmap, norm=norm)
plt.yticks(np.linspace(0,185,5,dtype=int),np.linspace(0,185,5,dtype=int)+1)
plt.ylabel('Residue index')
plt.xlabel('Time (ns)')
```
